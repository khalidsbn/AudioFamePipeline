### EDA

#### Numerical features

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.histplot(df['track_popularity'], bins=30, kde=True, color='green')
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

* **Inbalanced data, should balanced it in data processing by deleting 80% of rows where df.track_popularity == 0**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='danceability', y='energy', data=df, hue='track_popularity', palette='viridis', alpha=0.7)
plt.title('Danceability vs Energy (Coloured by Popularity)')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.legend(title='Popularity', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
top_genres = df['playlist_genre'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='muted')
plt.title('Top 10 Genres by Number of Songs')
plt.xlabel('Number of Songs')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize=(10, 8))

numeric_cols = ['danceability', 'energy', 'loudness', 'speechiness',
               'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='playlist_genre', y='loudness', data=df, palette='pastel')
plt.title('Loudness Distribution by Playlist Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='playlist_subgenre', y='tempo', data=df, palette='Set2', inner='quartile')
plt.title('Tempo Distribution by Playlist Subgenre')
plt.xticks(rotation=45)
plt.show()
```

#### Strings features

```python
# Count the top 10 values in a string column (e.g., 'playlist_genre')
top_categories = df['playlist_genre'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_categories.values, y=top_categories.index, palette='Blues_r')
plt.title('Top 10 Playlist Genres')
plt.xlabel('Frequency')
plt.ylabel('Genre')
plt.show()
```

```python
from wordcloud import WordCloud

# Combine all titles into one string
text_data = " ".join(df['track_name'].dropna().astype(str))

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_data)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Common Words in Track Names')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='playlist_genre', order=df['playlist_genre'].value_counts().index, palette='muted')
plt.title('Distribution of Playlist Genres')
plt.xticks(rotation=45)
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd

# Create a crosstab between two string columns
cross_tab = pd.crosstab(df['playlist_genre'], df['playlist_subgenre'])

plt.figure(figsize=(12, 8))
sns.heatmap(cross_tab, cmap='YlGnBu', linewidths=0.5)
plt.title('Relationship Between Genre and Subgenre')
plt.ylabel('Playlist Genre')
plt.xlabel('Playlist Subgenre')
plt.show()
```

```python
top_genres = df['playlist_genre'].value_counts().head(5)

plt.figure(figsize=(8, 8))
plt.pie(top_genres, labels=top_genres.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Top 5 Playlist Genres Distribution')
plt.show()

```

```python
df['title_length'] = df['track_name'].astype(str).apply(len)

plt.figure(figsize=(8, 6))
sns.histplot(df['title_length'], bins=30, kde=True, color='coral')
plt.title('Distribution of Track Name Lengths')
plt.xlabel('Title Length (characters)')
plt.ylabel('Frequency')
plt.show()
```

```python
unique_genres = df['playlist_genre'].nunique()
print(f"Number of unique playlist genres: {unique_genres}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histogram
plt.figure(figsize=(8,5))
sns.histplot(df['instrumentalness'], bins=30, kde=True, color='blue')

# Labels and title
plt.xlabel('Instrumentalness')
plt.ylabel('Count')
plt.title('Distribution of Instrumentalness')
plt.show()
```

Instrumentalness measures how likely a track is to be instrumental (without vocals). It ranges from 0 to 1:

* 0 → Mostly vocals
* 1 → Purely instrumental

Higher values indicate a greater likelihood that the track has no singing or spoken words.
